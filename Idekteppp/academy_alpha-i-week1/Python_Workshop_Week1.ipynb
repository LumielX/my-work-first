## 🔴 Part 8 — try / except / finally

**สิ่งที่ต้องรู้:**
```python
try:
    # โค้ดที่อาจเกิด Error
except ZeroDivisionError:
    # ทำงานเมื่อเกิด Error ประเภทนี้
except:
    # จับ Error ทุกประเภท
finally:
    # ทำงานเสมอ ไม่ว่าจะ Error หรือไม่
```

| คำสั่ง | ความหมาย |
|--------|----------|
| `try` | โค้ดที่อาจเกิด Error |
| `except` | ทำงานเมื่อเกิด Error |
| `finally` | ทำงานเสมอ |

### ข้อ 8.1 — ป้องกัน ZeroDivisionError
# Example
เขียนโค้ดหาร `10 / 0` แต่ดักจับ Error  
แสดงผล:
- **"Error: หารด้วยศูนย์ไม่ได้"**
- **"จบการทำงาน"** (ใน finally)

````python
# Example
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: หารด้วยศูนย์ไม่ได้")
finally:
    print("จบการทำงาน")
````

### ข้อ 8.2 — ฟังก์ชัน + try/except 
# Example

สร้างฟังก์ชัน `safe_divide(a, b)` ที่:
- ถ้าหารได้ → return ผลหาร
- ถ้า b=0 → return string `"ไม่สามารถหารด้วย 0"`

เรียกทดสอบ:
```
print(f"10 / 2 = {safe_divide(10, 2)}")   → "10 / 2 = 5.0"
print(f"10 / 0 = {safe_divide(10, 0)}")   → "10 / 0 = ไม่สามารถหารด้วย 0"
```

````python
# Example
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "ไม่สามารถหารด้วย 0"

print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")
````
